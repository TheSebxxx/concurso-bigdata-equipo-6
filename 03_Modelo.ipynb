```python
import pandas as pd
import matplotlib.pyplot as plt

# Ajusta la ruta si el CSV está en otra carpeta (recomendado: data/raw/)
RUTA_CSV = "../Reporte_Hurto_por_Modalidades_Policía_Nacional_20260619.csv"
df = pd.read_csv(RUTA_CSV, sep=",", encoding="utf-8")
df.shape
```

## 1. Estructura general del dataset

```python
df.info()
```

```python
df.head()
```

## 2. Calidad de datos: nulos y duplicados
Esto responde directamente a las dimensiones de **completitud** y **unicidad** que se documentan en `docs/conclusiones.md`.

```python
# Nulos por columna
df.isnull().sum().sort_values(ascending=False)
```

```python
# Duplicados exactos (mismo municipio + fecha + modalidad + cantidad)
cols_clave = [c for c in ["DEPARTAMENTO", "MUNICIPIO", "FECHA HECHO", "TIPO DE HURTO", "CANTIDAD"] if c in df.columns]
n_duplicados = df.duplicated(subset=cols_clave).sum()
print(f"Registros duplicados detectados: {n_duplicados} de {len(df)}")
```

## 3. Municipios y departamentos con mayor concentración de hurtos

```python
top_departamentos = df["DEPARTAMENTO"].value_counts().head(10)
top_departamentos
```

```python
top_municipios = df["MUNICIPIO"].value_counts().head(10)
top_municipios
```

```python
top_municipios.plot(kind="barh", figsize=(7,5), title="Top 10 municipios con más hurtos")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

## 4. Modalidad predominante: residencial vs. comercial

```python
df["TIPO DE HURTO"].value_counts()
```

```python
df["TIPO DE HURTO"].value_counts().plot(kind="pie", autopct="%1.1f%%", figsize=(5,5))
plt.ylabel("")
plt.title("Distribución por modalidad de hurto")
plt.show()
```

## 5. Armas y medios más frecuentes

```python
df["ARMAS MEDIOS"].value_counts().head(10)
```

## 6. Estacionalidad: tendencia mensual

```python
df["FECHA HECHO"] = pd.to_datetime(df["FECHA HECHO"], errors="coerce")
serie_mensual = df.groupby(df["FECHA HECHO"].dt.to_period("M"))["CANTIDAD"].sum()
serie_mensual
```

```python
serie_mensual.plot(kind="line", marker="o", figsize=(9,4), title="Hurtos totales por mes")
plt.ylabel("Cantidad de hurtos")
plt.tight_layout()
plt.show()
```
